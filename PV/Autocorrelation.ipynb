```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from datetime import timedelta
from dateutil.parser import parse
from datetime import datetime
from pandas.tseries.offsets import Hour, Minute
from pandas.tseries.offsets import Day, MonthEnd
import pytz
from pandas.tseries.offsets import Hour
from pandas.tseries.frequencies import to_offset
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
#df = pd.read_csv('Wind_deseason.csv')
df = pd.read_csv('PV_deseason_15.csv')
lags=len(df['310_PV_1'])
print(lags)
```

```python
sns.set_style("dark")

acf_plot = plot_acf(df['310_PV_1'], lags=120)
plt.xlabel("Time lags (k)",fontsize=15)
plt.ylabel("Autocorrelation",fontsize=15)
plt.savefig('PV_autocorrelation.svg')
```

```python
acf_plot = plot_acf(df['310_PV_1'], lags=20)
```

```python
sns.set_style("dark")
plot_pacf(df['310_PV_1'], lags=120, alpha = 0.05)
plt.xlabel("Time lags (k)",fontsize=15)
plt.ylabel("Autocorrelation",fontsize=15)
plt.savefig('PV_partialautocorrelation.png')
```
